```python
import pickle
import os
import pandas as pd
import numpy as np

# ì¤‘ìš”ë„ ëˆ„ì 
feature_importance_total = {}

for fold in range(5):
    file_path = f'XGBoost/importance_fold{fold}.pkl'
    if os.path.exists(file_path):
        with open(file_path, 'rb') as f:
            importance_dict = pickle.load(f)

        for k, v in importance_dict.items():
            feature_importance_total[k] = feature_importance_total.get(k, 0) + v

# F1 í‰ê·  ê³„ì‚°
f1_scores = []

for fold in range(5):
    file_path = f'XGBoost/f1_fold{fold}.txt'
    if os.path.exists(file_path):
        with open(file_path, 'r') as f:
            f1_scores.append(float(f.read()))

# ì¤‘ìš”ë„ ì •ë¦¬
importances = pd.Series(feature_importance_total).sort_values(ascending=False)

# âœ… ì¤‘ìš”ë„ ì •ê·œí™” (ì „ì²´ í•© ëŒ€ë¹„ ë¹„ìœ¨ë¡œ)
importances_normalized = importances / importances.sum()

# ìƒìœ„ 30ê°œ ì¶œë ¥
# "ê° í”¼ì²˜ê°€ ì „ì²´ ëª¨ë¸ ì„±ëŠ¥ì— ì–¼ë§ˆë‚˜ ê¸°ì—¬í–ˆëŠ”ê°€"
print("\nğŸ“Š ì •ê·œí™”ëœ ì¤‘ìš”ë„ ìƒìœ„ í”¼ì²˜:")
print(importances_normalized.head(30))

# ëˆ„ì  ë¹„ìœ¨ í™•ì¸ (ìƒìœ„ ëª‡ %ì¸ì§€ íŒë‹¨ ê°€ëŠ¥)
# "ìƒìœ„ Nê°œ í”¼ì²˜ê°€ ì „ì²´ ëª¨ë¸ ì„¤ëª…ë ¥ì˜ ëª‡ %ë¥¼ ì°¨ì§€í•˜ëŠ”ê°€"
cumsum = importances_normalized.cumsum()
print("\nğŸ“ˆ ìƒìœ„ 30ê°œ ëˆ„ì  ë¹„ìœ¨:")
print(cumsum.head(30))

# F1 í‰ê·  ì¶œë ¥
print(f"\nğŸ¯ í‰ê·  f1_micro score: {np.mean(f1_scores):.4f}")
```
