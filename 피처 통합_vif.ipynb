```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

def calculate_vif(df):
    X_const = add_constant(df)  # ìƒìˆ˜í•­ ì¶”ê°€
    vif_data = pd.DataFrame()
    vif_data["feature"] = df.columns
    vif_data["VIF"] = [
        variance_inflation_factor(X_const.values, i + 1)  # +1: ìƒìˆ˜í•­ ì œì™¸
        for i in range(df.shape[1])
    ]
    return vif_data.sort_values(by="VIF", ascending=False)
```

```python
# ê²°ì¸¡ì¹˜ ì œê±°ê°€ ì™„ë£Œë˜ì—ˆìœ¼ë¯€ë¡œ ë°”ë¡œ ê³„ì‚°
vif_result = calculate_vif(X_selected)

# ê²°ê³¼ í™•ì¸
display(vif_result)
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
import warnings
from catboost import CatBoostClassifier

warnings.filterwarnings('ignore')

# # êµì°¨ê²€ì¦ í•¨ìˆ˜ ìˆ˜ì •
# def evaluate_cv_score(X_input, y_input, cv=5):
#     model = CatBoostClassifier(
#         loss_function="MultiClass",
#         eval_metric="MultiClass",
#         task_type="GPU",
#         learning_rate=0.01,
#         iterations=70000,
#         l2_leaf_reg=50,
#         random_seed=42,
#         od_type="Iter",
#         depth=5,
#         border_count=64,
#         verbose=0
#     )
#     scores = cross_val_score(model, X_input, y_input, cv=cv, scoring='f1_micro')
#     return scores.mean()

# â–¶ ì´ˆê¸° ì„¤ì •
X_vif = X_selected.copy()
# y_vif = df_features.loc[X_vif.index, 'Segment_encoded']
iteration = 0

while True:
    # 1. ìƒìˆ˜í•­ ì¶”ê°€ í›„ VIF ê³„ì‚°
    X_const = sm.add_constant(X_vif)
    vif = pd.DataFrame()
    vif['ë³€ìˆ˜'] = X_const.columns
    vif['VIF'] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]
    vif = vif.sort_values(by='VIF', ascending=False)

    # 2. constëŠ” ì œì™¸í•˜ê³  ê°€ì¥ VIF ë†’ì€ ì»¬ëŸ¼ ì°¾ê¸°
    vif_no_const = vif[vif['ë³€ìˆ˜'] != 'const']
    top_vif = vif_no_const.iloc[0]

    print(f"[{iteration}] ğŸ” ìµœìƒìœ„ VIF ì»¬ëŸ¼: {top_vif['ë³€ìˆ˜']} (VIF={top_vif['VIF']:.2f})")

    # 3. ëª¨ë‘ ê¸°ì¤€ ì´í•˜ë¼ë©´ ì¢…ë£Œ
    if top_vif['VIF'] <= 5:
        print(f"âœ… ëª¨ë“  ì»¬ëŸ¼ì˜ VIFê°€ 5 ì´í•˜ì…ë‹ˆë‹¤. ì¢…ë£Œí•©ë‹ˆë‹¤.")
        break

    # # 4. ì„±ëŠ¥ ì¸¡ì •
    # f1 = evaluate_cv_score(X_vif, y_vif)
    # print(f"   í˜„ì¬ ì»¬ëŸ¼ ìˆ˜: {X_vif.shape[1]}, F1 Score: {f1:.4f}")

    # 5. ì»¬ëŸ¼ ì œê±°
    print(f"   ğŸ—‘ï¸ '{top_vif['ë³€ìˆ˜']}' ì œê±°\n")
    X_vif = X_vif.drop(columns=[top_vif['ë³€ìˆ˜']])
    iteration += 1
```
