```python
import pandas as pd
import gc
import numpy as np
```

```python
months = ["07", "08", "09", "10", "11", "12"]
monthly_list = []

for month_str in months:
    file_path = f"결측치 처리/2018{month_str}_train_1_회원정보_직장시도명행제거.csv"
    var_name = f"df{month_str}"  # ← 여기 이름을 var_name으로 맞춤

    try:
        df = pd.read_csv(file_path)
        globals()[var_name] = df
        monthly_list.append(df)
        print(f"✅ {var_name} 생성 완료 (파일: {file_path})")
    except FileNotFoundError:
        print(f"⚠️ 파일 없음: {file_path}")
```

```python
merged_df = pd.concat(monthly_list, ignore_index=True)
merged_df
```

```python
merged_df = merged_df.drop('Segment', axis=1)
```

```python
merged_df.to_csv('결측치 처리/병합/train_1_회원정보_직장시도명행제거.csv', index=False, encoding='utf-8-sig')
```
