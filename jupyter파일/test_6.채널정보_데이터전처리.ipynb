```python
import pandas as pd
import gc
```

```python
# csv 파일 데이터를 읽어온다.
df1 = pd.read_csv('to_csv_all/to_csv/test/6.채널정보/201807_test_채널정보.csv')
df2 = pd.read_csv('to_csv_all/to_csv/test/6.채널정보/201808_test_채널정보.csv')
df3 = pd.read_csv('to_csv_all/to_csv/test/6.채널정보/201809_test_채널정보.csv')
df4 = pd.read_csv('to_csv_all/to_csv/test/6.채널정보/201810_test_채널정보.csv')
df5 = pd.read_csv('to_csv_all/to_csv/test/6.채널정보/201811_test_채널정보.csv')
df6 = pd.read_csv('to_csv_all/to_csv/test/6.채널정보/201812_test_채널정보.csv')
```

```python
# 리스트로 모으기
dfs = [df1, df2, df3, df4, df5, df6]

# 그냥 concat
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
# 결측치 확인
combined_df.isna().sum()
```

```python
# 결측치 제거
df_test = combined_df.drop(columns=['OS구분코드'])
```

```python
# 결측치 여부 확인
df_test.isna().sum()
```

```python
## 데이터프레임을 넣고 column별 특성 및 결측값, 고유값들을 확인하는 함수를 작성해본다.
## 필수는 아니지만 전체적인 흐름을 파악하기 쉬워진다.

def resumetable(df, n):
    print(f'데이터셋 크기: {df.shape}')                                # 데이터프레임의 전체 크기(행, 열) 출력

    summary = pd.DataFrame(df.dtypes, columns=['데이터 타입'])         # 각 피처의 데이터 타입을 가져와 데이터프레임으로 생성
    summary = summary.reset_index()                                    # 인덱스를 초기화하여 컬럼으로 변환
    summary = summary.rename(columns={'index':'피처'})                 # 'index' 컬럼명을 '피처'로 변경
    summary['결측값 개수'] = df.isnull().sum().values                 # 각 피처의 결측값(null) 개수 계산
    summary['고유값 개수'] = df.nunique().values                      # 각 피처의 고유값 개수 계산
    summary['최솟값'] = df.min(numeric_only=True).reindex(df.columns).values           # 최솟값
    summary['최댓값'] = df.max(numeric_only=True).reindex(df.columns).values           # 최댓값

    for i in range(n):
        summary[f"{i+1}번째 값"] = df.iloc[i].values
    
    return summary                                                     # 요약 테이블 반환
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(resumetable(df_test, 1))
```
