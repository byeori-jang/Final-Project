```python
def show_segment_distribution(df, columns, segment_col='Segment'):
    """
    ê° ì»¬ëŸ¼ì˜ ê³ ìœ ê°’ë³„ë¡œ Segment ë¶„í¬ (ì¹´ìš´íŠ¸ì™€ ë¹„ìœ¨) ì¶œë ¥
    """
    for col in columns:
        print(f"\n\nğŸŸ¢ [ì»¬ëŸ¼: {col}]")
        
        # ê·¸ë£¹ë³„ ê°œìˆ˜
        count_table = df.groupby([col, segment_col]).size().unstack(fill_value=0)
        print("âœ… Count:")
        display(count_table)
        
        # ê·¸ë£¹ë³„ ë¹„ìœ¨ (row-wise ë¹„ìœ¨, ë°±ë¶„ìœ¨ë¡œ ë³€í™˜)
        ratio_table = count_table.div(count_table.sum(axis=0), axis=1) * 100
        print("âœ… Ratio (%):")
        display(ratio_table.round(4))  # ë³´ê¸° ì¢‹ê²Œ ì†Œìˆ˜ì  ë‘˜ì§¸ìë¦¬ê¹Œì§€
```

```python
show_segment_distribution(selected_df, cols_to_drop)
```

```python
import pandas as pd

def evaluate_segment_usefulness(df, columns, segment_col='Segment', rare_thresh=100):
    results = []

    for col in columns:
        value_counts = df[col].value_counts()
        rare_values = value_counts[value_counts < rare_thresh].index

        # Segment count & ratio
        count_table = df.groupby([col, segment_col]).size().unstack(fill_value=0)
        ratio_table = count_table.div(count_table.sum(axis=1), axis=0)

        # Segmentë³„ ìµœëŒ€-ìµœì†Œ ë¹„ìœ¨ ì°¨ì´
        max_gap_per_segment = ratio_table.max() - ratio_table.min()
        total_max_gap = max_gap_per_segment.max()

        # Segment ì ë¦¼ë„ (ê°’ë³„ ìµœê³  ë¹„ìœ¨ì˜ í‰ê· )
        max_ratios = ratio_table.max(axis=1)
        mean_peak_ratio = max_ratios.mean()

        # í¬ê·€ê°’ ë¹„ìœ¨
        rare_ratio = df[col].isin(rare_values).mean()

        # ìœ ì˜ë¯¸ë„ íŒì •
        if total_max_gap >= 0.4 and mean_peak_ratio >= 0.75:
            judgment = 'ë†’ìŒ'
        elif total_max_gap >= 0.2 and mean_peak_ratio >= 0.6:
            judgment = 'ì¤‘ê°„'
        else:
            judgment = 'ë‚®ìŒ'


        results.append({
            'ì»¬ëŸ¼': col,
            'ê°’ ìˆ˜': df[col].nunique(),
            'ìµœëŒ€ë¹„ìœ¨ì°¨(%)': round(total_max_gap * 100, 1),
            'ì ë¦¼ë„í‰ê· (%)': round(mean_peak_ratio * 100, 1),
            'í¬ê·€ê°’ë¹„ì¤‘(%)': round(rare_ratio * 100, 2),
            'ìœ ì˜ë¯¸ì„±': judgment
        })

    result_df = pd.DataFrame(results).sort_values(by='ìœ ì˜ë¯¸ì„±', ascending=False)
    return result_df
```

```python
cols_to_check = cols_to_drop
result = evaluate_segment_usefulness(selected_df, cols_to_check)
display(result)
```
