```python
# 1. 필수 라이브러리 불러오기
import pandas as pd

# 2. 파일 불러오기 (경로는 너의 파일 위치에 맞게)
file_path = "C:/Users/PC/Desktop/final/open/train/7.마케팅정보/201807_train_마케팅정보.parquet"
df = pd.read_parquet(file_path)

# 3. 요약 함수 정의
def resumetable(df, n=5):
    summary = pd.DataFrame(df.dtypes, columns=['데이터 타입']).reset_index()
    summary = summary.rename(columns={'index': '피처'})
    summary['결측값 개수'] = df.isnull().sum().values
    summary['고유값 개수'] = df.nunique().values

    for i in range(n):
        if i < len(df):
            summary[f'{i+1}번째 값'] = df.iloc[i].values
        else:
            summary[f'{i+1}번째 값'] = [None]*df.shape[1]
    
    return summary

# 4. 결과 출력 (행 생략 없이 보기 좋게)
with pd.option_context('display.max_rows', None):
    display(resumetable(df, 5))
```

```python
import pandas as pd

# 1. 데이터 불러오기 (.parquet 파일)
file_path = "C:/Users/PC/Desktop/final/open/train/7.마케팅정보/201807_train_마케팅정보.parquet"
df = pd.read_parquet(file_path)

# 2. 열 통계 요약 함수 정의
def column_summary(df):
    summary = pd.DataFrame()
    summary['타입'] = df.dtypes
    summary['결측치'] = df.isnull().sum()
    summary['고유값'] = df.nunique()
    
    # 숫자형 데이터만 가능한 항목들 (오류 방지용)
    summary['최댓값'] = df.max(numeric_only=True)
    summary['최솟값'] = df.min(numeric_only=True)
    
    # 최빈값은 모든 타입에 대해 처리 가능하나, mode가 여러 개일 수 있음 → 첫 번째만 사용
    summary['최빈값'] = df.mode().iloc[0]
    
    summary = summary.reset_index().rename(columns={'index': '컬럼명'})
    return summary

# 3. 요약 실행
summary_df = column_summary(df)

# 4. 엑셀로 내보내기 (엑셀 템플릿과 결합하려면 openpyxl 필요)
summary_df.to_excel("컬럼요약_마케팅정보.xlsx", index=False)
```
