```python
def show_segment_distribution(df, columns, segment_col='Segment'):
    """
    ê° ì»¬ëŸ¼ì˜ ê³ ìœ ê°’ë³„ë¡œ Segment ë¶„í¬ (ì¹´ìš´íŠ¸ì™€ ë¹„ìœ¨) ì¶œë ¥
    """
    for col in columns:
        print(f"\n\nğŸŸ¢ [ì»¬ëŸ¼: {col}]")
        
        # ê·¸ë£¹ë³„ ê°œìˆ˜
        count_table = df.groupby([col, segment_col]).size().unstack(fill_value=0)
        print("âœ… Count:")
        display(count_table)
        
        # ê·¸ë£¹ë³„ ë¹„ìœ¨ (row-wise ë¹„ìœ¨, ë°±ë¶„ìœ¨ë¡œ ë³€í™˜)
        ratio_table = count_table.div(count_table.sum(axis=1), axis=0) * 100
        print("âœ… Ratio (%):")
        display(ratio_table.round(4))  # ë³´ê¸° ì¢‹ê²Œ ì†Œìˆ˜ì  ë‘˜ì§¸ìë¦¬ê¹Œì§€

        print(f"{col} ê°’ë³„ ë¹„ì¤‘:\n{df[col].value_counts(normalize=True)}")
```

```python
show_segment_distribution(selected_df, cols_to_drop)
```

```python
def show_segment_id_counts(df, columns, segment_col='Segment', id_col='ID'):
    """
    ê° ì»¬ëŸ¼ì˜ ê³ ìœ ê°’ë³„ë¡œ Segment ë¶„í¬ (IDì˜ ê³ ìœ  ê°œìˆ˜ ê¸°ì¤€) ì¶œë ¥
    """
    for col in columns:
        print(f"\n\nğŸŸ¢ [ì»¬ëŸ¼: {col}]")
        
        # ID ê¸°ì¤€ ê³ ìœ  ê°œìˆ˜ ê³„ì‚° (pivot í˜•íƒœ)
        count_table = (
            df.groupby([col, segment_col])[id_col]
            .nunique()
            .unstack(fill_value=0)
        )
        
        print("âœ… Unique ID Count:")
        display(count_table)
        
        # ë¹„ìœ¨ (row-wise ë¹„ìœ¨)
        ratio_table = count_table.div(count_table.sum(axis=0), axis=1) * 100
        print("âœ… Ratio (%):")
        display(ratio_table.round(4))

        print(f"{col} ê°’ë³„ ë¹„ì¤‘:\n{df[col].value_counts(normalize=True)}")
```

```python
show_segment_id_counts(selected_df, cols_to_drop)
```
