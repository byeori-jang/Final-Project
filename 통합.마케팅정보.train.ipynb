```python
import pandas as pd

# CSV 파일 불러오기
file_path = "train_7_마케팅정보.csv"
df = pd.read_csv(file_path)

cols_to_drop = []

for col in df.columns:
    unique_vals = df[col].dropna().unique()
    val_counts = df[col].value_counts(dropna=True)

    # 조건 1: 고유값이 1개뿐인 컬럼
    if len(unique_vals) == 1:
        cols_to_drop.append(col)

    # 조건 2: 고유값이 2개인데 한쪽은 10개 이하, 다른 쪽은 나머지 거의 전체
    elif len(unique_vals) == 2:
        values = val_counts.values
        if (values[0] <= 10 and values[1] > 100_000) or (values[1] <= 10 and values[0] > 100_000):
            cols_to_drop.append(col)

# 컬럼 제거
df_cleaned = df.drop(columns=cols_to_drop)

# 결과 저장 (선택)
df_cleaned.to_csv("train_7_마케팅정보_unique.csv", index=False , encoding='utf-8-sig')

# 제거한 컬럼 목록 확인
print("제거된 컬럼:", cols_to_drop)
```

```python
import pandas as pd

# 이전에 컬럼 제거된 상태의 데이터 불러오기
df = pd.read_csv("train_7_마케팅정보_unique.csv")  # 방금 저장했던 파일 기준

# 새롭게 제거할 컬럼 리스트
cols_to_drop_98 = []

for col in df.columns:
    val_counts = df[col].value_counts(dropna=True)
    total = val_counts.sum()

    # 조건: 한 값이 98% 이상인 컬럼
    if len(val_counts) >= 2 and (val_counts.iloc[0] / total) >= 0.98:
        cols_to_drop_98.append(col)

# 제거 적용
df_cleaned_v2 = df.drop(columns=cols_to_drop_98)

# 파일 저장
df_cleaned_v2.to_csv("train_7_마케팅정보_98이상제거.csv", index=False , encoding='utf-8-sig')

# 확인 출력
print("98% 이상 편향된 컬럼 수:", len(cols_to_drop_98))
print("제거된 컬럼:", cols_to_drop_98)
```
