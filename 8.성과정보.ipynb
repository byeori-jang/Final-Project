```python
import pandas as pd

# ✅ 월별 파일 경로 설정
file_path = "C:/Users/PC/Desktop/final/open/train/8.성과정보/201807_train_성과정보.parquet"
df = pd.read_parquet(file_path)

# ✅ 요약 함수 정의
def column_summary(df):
    summary = pd.DataFrame()
    summary['타입'] = df.dtypes
    summary['결측치'] = df.isnull().sum()
    summary['고유값'] = df.nunique()
    summary['최댓값'] = df.max(numeric_only=True)
    summary['최솟값'] = df.min(numeric_only=True)
    summary['최빈값'] = df.mode().iloc[0]
    summary = summary.reset_index().rename(columns={'index': '컬럼명'})
    return summary

# ✅ 요약 실행 및 엑셀 저장
summary_df = column_summary(df)
summary_df.to_excel("C:/Users/PC/Desktop/201807_컬럼요약_성과정보.xlsx", index=False)
```
