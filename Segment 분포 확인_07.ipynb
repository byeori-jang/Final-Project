```python
def show_segment_distribution(df, columns, segment_col='Segment'):
    """
    Í∞Å Ïª¨ÎüºÏùò Í≥†Ïú†Í∞íÎ≥ÑÎ°ú Segment Î∂ÑÌè¨ (Ïπ¥Ïö¥Ìä∏ÏôÄ ÎπÑÏú®) Ï∂úÎ†•
    """
    for col in columns:
        print(f"\n\nüü¢ [Ïª¨Îüº: {col}]")
        
        # Í∑∏Î£πÎ≥Ñ Í∞úÏàò
        count_table = df.groupby([col, segment_col]).size().unstack(fill_value=0)
        print("‚úÖ Count:")
        display(count_table)
        
        # Í∑∏Î£πÎ≥Ñ ÎπÑÏú® (row-wise ÎπÑÏú®, Î∞±Î∂ÑÏú®Î°ú Î≥ÄÌôò)
        ratio_table = count_table.div(count_table.sum(axis=0), axis=1) * 100
        print("‚úÖ Ratio (%):")
        display(ratio_table.round(4))  # Î≥¥Í∏∞ Ï¢ãÍ≤å ÏÜåÏàòÏ†ê ÎëòÏß∏ÏûêÎ¶¨ÍπåÏßÄ
```

```python
import pandas as pd

def evaluate_segment_usefulness(df, columns, segment_col='Segment', rare_thresh=100):
    results = []

    for col in columns:
        value_counts = df[col].value_counts()
        rare_values = value_counts[value_counts < rare_thresh].index

        # Segment count & ratio
        count_table = df.groupby([col, segment_col]).size().unstack(fill_value=0)
        ratio_table = count_table.div(count_table.sum(axis=1), axis=0)

        # SegmentÎ≥Ñ ÏµúÎåÄ-ÏµúÏÜå ÎπÑÏú® Ï∞®Ïù¥
        max_gap_per_segment = ratio_table.max() - ratio_table.min()
        total_max_gap = max_gap_per_segment.max()

        # Segment Ïè†Î¶ºÎèÑ (Í∞íÎ≥Ñ ÏµúÍ≥† ÎπÑÏú®Ïùò ÌèâÍ∑†)
        max_ratios = ratio_table.max(axis=1)
        mean_peak_ratio = max_ratios.mean()

        # Ìù¨Í∑ÄÍ∞í ÎπÑÏú®
        rare_ratio = df[col].isin(rare_values).mean()

        # Ïú†ÏùòÎØ∏ÎèÑ ÌåêÏ†ï
        if total_max_gap >= 0.4 and mean_peak_ratio >= 0.75:
            judgment = 'ÎÜíÏùå'
        elif total_max_gap >= 0.2 and mean_peak_ratio >= 0.6:
            judgment = 'Ï§ëÍ∞Ñ'
        else:
            judgment = 'ÎÇÆÏùå'


        results.append({
            'Ïª¨Îüº': col,
            'Í∞í Ïàò': df[col].nunique(),
            'ÏµúÎåÄÎπÑÏú®Ï∞®(%)': round(total_max_gap * 100, 1),
            'Ïè†Î¶ºÎèÑÌèâÍ∑†(%)': round(mean_peak_ratio * 100, 1),
            'Ìù¨Í∑ÄÍ∞íÎπÑÏ§ë(%)': round(rare_ratio * 100, 2),
            'Ïú†ÏùòÎØ∏ÏÑ±': judgment
        })

    result_df = pd.DataFrame(results).sort_values(by='Ïú†ÏùòÎØ∏ÏÑ±', ascending=False)
    return result_df
```

```python
cols_to_check = cols_to_drop
result = evaluate_segment_usefulness(selected_df, cols_to_check)
display(result)
```

```python
def show_segment_id_counts(df, columns, segment_col='Segment', id_col='ID'):
    """
    Í∞Å Ïª¨ÎüºÏùò Í≥†Ïú†Í∞íÎ≥ÑÎ°ú Segment Î∂ÑÌè¨ (IDÏùò Í≥†Ïú† Í∞úÏàò Í∏∞Ï§Ä) Ï∂úÎ†•
    """
    for col in columns:
        print(f"\n\nüü¢ [Ïª¨Îüº: {col}]")
        
        # ID Í∏∞Ï§Ä Í≥†Ïú† Í∞úÏàò Í≥ÑÏÇ∞ (pivot ÌòïÌÉú)
        count_table = (
            df.groupby([col, segment_col])[id_col]
            .nunique()
            .unstack(fill_value=0)
        )
        
        print("‚úÖ Unique ID Count:")
        display(count_table)
        
        # ÎπÑÏú® (row-wise ÎπÑÏú®)
        ratio_table = count_table.div(count_table.sum(axis=0), axis=1) * 100
        print("‚úÖ Ratio (%):")
        display(ratio_table.round(4))

        print(f"{col} Í∞íÎ≥Ñ ÎπÑÏ§ë:\n{df[col].value_counts(normalize=True)*100}")
```
