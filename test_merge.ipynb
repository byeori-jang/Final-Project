```python
import pandas as pd
import gc
import numpy as np
```

```python
base_path = 'open\\test\\2.신용정보\\'

months = ["07", "08", "09", "10", "11", "12"]
monthly_list = []

for month_str in months:
    file_path = f"{base_path}2018{month_str}_test_신용정보.parquet"
    var_name = f"df{month_str}"  # ← 여기 이름을 var_name으로 맞춤

    try:
        df = pd.read_parquet(file_path, engine = "pyarrow")
        globals()[var_name] = df
        monthly_list.append(df)
        print(f"✅ {var_name} 생성 완료 (파일: {file_path})")
    except FileNotFoundError:
        print(f"⚠️ 파일 없음: {file_path}")
```

```python
merged_df = pd.concat(monthly_list, ignore_index=True)
merged_df
```

```python
merged_df.to_csv('test_2_신용정보_병합.csv', index=False, encoding='utf-8-sig')
```
