```python
import pandas as pd
import gc
import numpy as np
```

```python
base_path = 'open/train/1.회원정보/'

months = ["07", "08", "09", "10", "11", "12"]
monthly_list = []

for month_str in months:
    file_path = f"{base_path}2018{month_str}_train_회원정보.parquet"
    var_name = f"df{month_str}"  # ← 여기 이름을 var_name으로 맞춤

    try:
        df = pd.read_parquet(file_path, engine = "pyarrow")
        # df = pd.read_csv(file_path)
        globals()[var_name] = df
        monthly_list.append(df)
        print(f"✅ {var_name} 생성 완료 (파일: {file_path})")
    except FileNotFoundError:
        print(f"⚠️ 파일 없음: {file_path}")
```

```python
merged_df = pd.concat(monthly_list, ignore_index=True)
merged_df
```

```python
seg = merged_df[['ID', 'Segment']]
seg
```

```python
seg.to_csv('결측치 처리/Segment/Segment_병합.csv', index=False, encoding='utf-8-sig')
```
